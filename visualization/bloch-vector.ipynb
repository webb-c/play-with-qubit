## Qiskit Visualization: bloch-vector

Quantum state vector를 나타내기 위해서 **Bloch sphere**라는 3차원 벡터공간에서의 unit space를 이용한다. 각각의 basis들의 amplitude들이 real number 또는 complex number임을 이용하여 polar form 그리고 cartesian form으로 나타낼 수 있다.

```python
from qiskit.visualization import plot_bloch_vector
from math import pi, cos, sin
```

### Polar form

다음과 같이 어떤 임의의 quantum state vector가 존재한다고 하자.
$$ \ket \psi = \alpha \ket 0 + \beta \ket 1$$ 
$$ \ket \psi = r_\alpha e^{i \theta_\alpha} \ket 0 + r_\beta e^{i \theta_\beta} \ket 1$$ 

이때, 두 basis state에 모두 곱해지는 **global phase** $e^{i\theta}$는 실제 물리적으로는 무시되는 값이다. 


따라서 $\ket 0$에 대한 phase를 $e^{i \theta_\alpha}$를 소거하여 $\ket 1$에 대한 **relative phase** $e^{i (\theta_\beta - \theta_\alpha)}$만 남길 수 있다.

$$ \ket \psi = r_\alpha \ket 0 + r_\beta e^{i (\theta_\beta - \theta_\alpha)} \ket 1$$ 

여기서 우리는 다음의 조건들을 지켜야한다. 
- $|\alpha|^2 + |\beta|^2 = 1$
- $\alpha$는 real value (positive)이며, relative phase를 가질 수도 있는 $\beta$는 complex number가 될 수도 있다.

이를 모두 지키는 표현으로 다음과 같은 형태를 사용할 수 있다. 
$$\ket \psi = \cos(\theta/2) + e^{i \phi} \sin(\theta/2)$$


그럼 우리는 이제 2개의 변수 $(\theta, \phi)$를 얻었고, 이를 Bloch Sphere에서 각각 다음과 같은 각도로 생각하면, 우리는 quantum state vector를 3차원에서의 unit space위에 시각화 할 수 있게 된다.

- $\theta$: z축 기준 $\ket 0$에서 떨어진 각도
- $\phi$: xy평면에서 $\ket +$에서 떨어진 각도

$$(\theta, \phi)$$

```python
# \theta = pi/2, \phi = pi/2
plot_bloch_vector([1, pi/2, pi/2], title="New Bloch Sphere (Polar form)", coord_type="spherical")
```

```python
# \theta = 3*pi/4, \phi = pi/4
plot_bloch_vector([1, 3*pi/4, pi/4], title="New Bloch Sphere (Polar form)", coord_type="spherical")
```

### Cartesian form
반면 우리는 각도 $\theta, \phi$를 알고 있기 때문에, 이를 이용하면 $x, y, z$ 평면에서 각각의 좌표를 계산할 수 있게 된다. 

가장 먼저 단순하게 얻을 수 있는 값은 바로 $z$ 좌표로서 이 벡터는 unit vector이므로 $r \cos \theta = \cos \theta$가 된다.

$$z = \cos \theta$$

반면, $x, y$좌표를 $xy$평면에 사용된 unit vector의 길이가 $\sin \theta$이므로 $r=\sin \theta$가 되어서 각각 다음 값을 가진다.

$$x = \sin \theta \cos \phi, \qquad y = \sin \theta \sin \phi$$


즉, 이 과정을 통해 $(\theta, \phi)$의 2가지 변수로 구성된 quantum state vector를 3차원 좌표계위의 점으로 변환할 수 있다. 이 값을 사용해서도 마찬가지로 시각화가 가능하다.
$$(x, y, z)$$

```python
# \theta = pi/2, \phi = pi/2

theta = pi/2
phi = pi/2

plot_bloch_vector([sin(theta)*cos(phi), sin(theta)*sin(phi), cos(theta)], title="New Bloch Sphere (Cartesian form)", coord_type="Cartesian")
```

```python
# \theta = 3*pi/4, \phi = pi/4

theta = 3*pi/4
phi = pi/4

plot_bloch_vector([sin(theta)*cos(phi), sin(theta)*sin(phi), cos(theta)], title="New Bloch Sphere (Cartesian form)", coord_type="Cartesian")
```
